### #Regression Algorithm

```
#X = tesla.drop("Close", axis = "columns")
X = np.array(tesla.index).reshape(-1,1)
y = tesla['Close']
print("shape of X = ", X.shape)
print("shape of y = ", y.shape)
```

```
y.head()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
```

## #Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train= sc.transform(X_train)
X_test = sc.transform(X_test)
```

```
 # y = m * x + c
 from sklearn.linear_model import LinearRegression
 lr = LinearRegression()
```

```
lr.fit(X_train,y_train)
```

```
plt.scatter(X_test, y_test)
plt.plot(X_test, lr.predict(X_test), color = "r")
```

### Model Accuracy Eveloution

```
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, lr.predict(X_test))
rmse = np.sqrt(mse)
print("MSE = ", mse)
print("RMSE = ", rmse)
```

### **LSTM Algorithm**

```
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```
tesla.head()
```

```
tesla1 = tesla.iloc[:2390,:]
tesla1.tail()
```

```
tesla2 = tesla.iloc[2390:2417,:]
tesla2.head()
```

```
traindata = tesla1.iloc[:,4:5].values
traindata
```

```
sc = MinMaxScaler(feature_range=(0,1))
traindata = sc.fit_transform(traindata)
traindata.shape
```

```
X1_train = []
y1_train = []

for i in range (60,2390):
  X1_train.append(traindata[i-60:i,0])
  y1_train.append(traindata[i,0])


X1_train, y1_train = np.array(X1_train), np.array(y1_train)
```

```
#Reshape
X1_train = np.reshape(X1_train, (X1_train.shape[0], X1_train.shape[1], 1))
```

```
#Initializing the RNN
model = Sequential()
```

```
#Adding the first LSTM Layer and some Dropout regularisation
model.add(LSTM(units = 100, return_sequences = True, input_shape = (X1_train.shape[1], 1)))
model.add(Dropout(0.2))

#Adding the second LSTM Layer and some Dropout regularisation
model.add(LSTM(units = 100, return_sequences = True))
model.add(Dropout(0.2))

#Adding the third LSTM Layer and some Dropout regularisation
model.add(LSTM(units = 100, return_sequences = True))
model.add(Dropout(0.2))

#Adding the third LSTM Layer and some Dropout regularisation
model.add(LSTM(units = 100))
model.add(Dropout(0.2))

#Adding the output Layer
model.add(Dense(units = 1))
```

```
#Compiling the RNN
model.compile(optimizer='adam', loss='mean_squared_error')

#Fitting the RNN to training set
model.fit(X1_train, y1_train, epochs = 20, batch_size =32)
```

### Testing

```
tesla2.head()
```

```
traindata1 = tesla2.iloc[:,4:5].values
```

```
dataset_total = pd.concat((tesla1['Close'], tesla2['Close']), axis = 0)
```

```
inputs = dataset_total[len(dataset_total) - len(tesla2) - 60:].values
```

```
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)
```

```
X1_test = []

for i in range (60,86):
  X1_test.append(inputs[i-60:i,0])


X1_test = np.array(X1_test)

#Reshape
X1_test = np.reshape(X1_test, (X1_test.shape[0], X1_test.shape[1], 1))
```

```
predict_tesla_stock_price = model.predict(X1_test)
```

```
predict_tesla_stock_price = sc.inverse_transform(predict_tesla_stock_price)
```

```
plt.plot(traindata1, color ='red', label = 'Real Tesla Stock Price')
plt.plot(predict_tesla_stock_price, color ='blue', label = 'Predicted Tesla Stock Price')
plt.xlabel("Time")
plt.ylabel("Tesla Stock Price")
plt.title("Tesla Stock Price Prediction")
plt.legend()
plt.show()
```
